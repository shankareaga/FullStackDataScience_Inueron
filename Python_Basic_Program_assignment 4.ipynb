###  Question 1
1.	Write a Python Program to Find the Factorial of a Number?
The factorial of a number is the product of all the integers from 1 to that number. For example, 
the factorial of 6 is 1*2*3*4*5*6 = 720 . 
Factorial is not defined for negative numbers, and the factorial of zero is one, 0!

```python
Input_Num = int(input("Enter number to find factorial"))
Fact_Num = 1

if Input_Num > 0:
    for i in range(2,Input_Num+1):
        Fact_Num*=i
    print("The facotrial of {} is equal to {}.".format(Input_Num, Fact_Num))  
elif Input_Num == 0:
    print("The factorial of zero is equal to 1")
else:
    print("The factorial is not defined for negative numbers")
```

```python
# Question 2
#2.	Write a Python Program to Display the multiplication Table?
# Enter a number to create a multiplication table from 1 and upto number 
Number = int(input("Enter the positive number for creating multiplication table: "))
for i in range (1, Number+1):
    print("Multiplication table for {} \n".format(i))
    for j in range(1, 11):
        print("{} * {} = {}".format(i, j, i*j))
    print("\n")
```

### 3.	Write a Python Program to Print the Fibonacci sequence?
In mathematics, the Fibonacci numbers, commonly denoted Fn, form a sequence, called the Fibonacci sequence, 
such that each number is the sum of the two preceding ones, starting from 0 and 1. 

```python
# define first two number zero and one
F0 = 0
F1 = 1
F_squence = [F0,F1]
F_number = int(input("Enter the number to create Fibonacci sequence : "))
for i in range(2, F_number):
    F_squence.append(F0+F1)
    F0, F1 = F1, F0+F1
print(F_squence)
```

```python
# Question 4
# 4.	Write a Python Program to Check Armstrong Number?
# An Armstrong Number is a number that when you raise each digit of the number to the power of the number of digits
# and then add them up you get the original number.
# a is a Armstrong Number if a=(a1)**n+(a2)**n...(an)**n

IsArmstrong_num = int(input("Enter the number to find whether it is a Armstrong or not:"))

# let say b = base which is equal to total digits of the input number
base = len(str(IsArmstrong_num))
sum_digits = 0
# split the number into digit
for i in str(IsArmstrong_num):
    sum_digits += int(i)**base

if sum_digits == IsArmstrong_num:
    print("The entered number is a Armstrong number {}".format(IsArmstrong_num))
else:
    print("The entered number {} is not a Armstrong number.".format(IsArmstrong_num))
```
