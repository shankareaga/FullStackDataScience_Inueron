```python
# Question 1
# 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?
 # Enter a number to find whether it is a positive , negative or zero
    
Input_Num = int(input("Enter a number "))

if Input_Num == 0:
    print("The entered number is zero")
elif Input_Num < 0:
    print("The entered number is negative")
else:
    print("The entered number is positive")
```

```python
# Question 2
# 2.	Write a Python Program to Check if a Number is Odd or Even?
Input_Num = int(input("Enter a number "))

if Input_Num % 2 == 0:
    print("The entered number is even")
else:
    print("The entered number is odd")
```

```python
# Question 3
# 3.	Write a Python Program to Check Leap Year?

Input_Year = int(input("Enter year number in yyyy format "))

if Input_Year % 4 == 0:
    print("The entered year {} is not a leap year".format(Input_Year))
else:
    print("The entered year {} is a leap year".format(Input_Year))
```

```python
# Question 4
# 4.	Write a Python Program to Check Prime Number?
# A prime number is a number which is divisible by 1 and by itself. Hence No need to check whether number is 
# divisible by 1 and itself , which is obviously the case for any prime number

Is_Prime = int(input("Enter the number to find whether it is a prime or not"))
               
for i in range(2,Is_Prime):
    if Is_Prime % i == 0:
       print(Is_Prime, "is not a prime number.")
       break
    else:
        print(Is_Prime, "is a prime number.")
        break
```

```python
# Question 5
# 5.	Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

for i in range(2,10000):
    for j in range(2,i):
        if i % j == 0:
            print(i, "not a prime number")
            break
    else:
        print(i, "prime number")
        
```
